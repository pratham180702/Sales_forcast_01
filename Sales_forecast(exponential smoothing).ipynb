```python
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df = pd.read_csv('train.csv', index_col = "Day")
# df = df.asfreq("D")
```

```python
model = ExponentialSmoothing(endog = df.GrocerySales).fit()
```

```python
predictions = model.forecast(steps =  31)
df['GrocerySales'].plot(figsize= (10,7))
predictions.plot()
```

```python
predictions_df = predictions.to_frame()
predictions_df
```
